## Task #6

```python
#Converting the TXN_DATE_TIME column into separate columns with year, month & date of the transactions
t = mdf1[['TXN_DATE_TIME','CLAIM_AMOUNT','INSURANCE_TYPE']]
t['TXN_DATE_TIME'] = pd.to_datetime(t['TXN_DATE_TIME'], dayfirst=True)
t['year']= t['TXN_DATE_TIME'].dt.year
t['month']= t['TXN_DATE_TIME'].dt.month
t['day']= t['TXN_DATE_TIME'].dt.day
t
```

```python
#Creating last_month_df containing transactions of only the last month of dataframe ie. 06/2021.
last_month_df = t[(t['month'] == 6) & (t['day'] >=1) & (t['year'] == 2021)]
```

```python
#Filtering last_month_df for same type of insurance
Mobile = last_month_df[last_month_df['INSURANCE_TYPE'] == 'Mobile']

#Calculating avg 
avg_Mobile = Mobile['CLAIM_AMOUNT'].sum()/len(Mobile)
Mobile['Avg_M'] = avg_Mobile

#Creating a column "CD" containing Claim Deviation by dividing avg of claim amount by current claim amount
Mobile['CD'] = Mobile['Avg_M']/Mobile['CLAIM_AMOUNT']

#Creating a column 'Final_CD' contaiing 1 if CD < 5 and 0 if not. 
Mobile['Final_CD'] = np.where(Mobile['CD'] < .5, 1,0)

#Repeating these steps for all insurance types

Property = last_month_df[last_month_df['INSURANCE_TYPE'] == 'Property']
avg_P = Property['CLAIM_AMOUNT'].sum()/len(Property)
Property['Avg_P'] = avg_P
Property['CD'] = Property['Avg_P']/Property['CLAIM_AMOUNT']
Property['Final_CD'] = np.where(Property['CD'] < .5, 1,0)

Health = last_month_df[last_month_df['INSURANCE_TYPE'] == 'Health']
avg_H = Health['CLAIM_AMOUNT'].sum()/len(Health)
Health['Avg_H'] = avg_H
Health['CD'] = Health['Avg_H']/Health['CLAIM_AMOUNT']
Health['Final_CD'] = np.where(Health['CD'] < .5, 1,0)

Motor = last_month_df[last_month_df['INSURANCE_TYPE'] == 'Motor']
avg_motor = Motor['CLAIM_AMOUNT'].sum()/len(Motor)
Motor['Avg_motor'] = avg_motor
Motor['CD'] = Motor['Avg_motor']/Motor['CLAIM_AMOUNT']
Motor['Final_CD'] = np.where(Motor['CD'] < .5, 1,0)

Travel = last_month_df[last_month_df['INSURANCE_TYPE'] == 'Travel']
avg_T = Travel['CLAIM_AMOUNT'].sum()/len(Travel)
Travel['Avg_T'] = avg_T
Travel['CD'] = Travel['Avg_T']/Travel['CLAIM_AMOUNT']
Travel['Final_CD'] = np.where(Travel['CD'] < .5, 1,0)

Life = last_month_df[last_month_df['INSURANCE_TYPE'] == 'Life']
avg_L = Life['CLAIM_AMOUNT'].sum()/len(Life)
Life['Avg_L'] = avg_L
Life['CD'] = Life['Avg_L']/Life['CLAIM_AMOUNT']
Life['Final_CD'] = np.where(Life['CD'] < .5, 1,0)
```

```python
#Combining all the above-made dataframes to make a final df containig CD and Final_CD values of all Insurance Types. 
lst = [Mobile,Property,Health,Life,Motor,Travel]
final_cd_df = pd.concat(lst)
#Answer
final_cd_df
```

## Task #8

```python
#Calculating the Change in premium amounts of Motor and Life insurance after marginal premium increase by 2%. 
am = mdf[['PREMIUM_AMOUNT','INSURANCE_TYPE']]
m = am[(am['INSURANCE_TYPE'] == 'Motor') | (am['INSURANCE_TYPE'] == 'Life')]
ml = m['PREMIUM_AMOUNT']*.02 + m['PREMIUM_AMOUNT']
change_ml = ml.sum() - m['PREMIUM_AMOUNT'].sum()
change_ml
```

```python
#Calculating the Change in premium amounts of Property and Health insurance after premium increase by 7%
p = am[(am['INSURANCE_TYPE'] == 'Property') | (am['INSURANCE_TYPE'] == 'Health')]
propPh = p['PREMIUM_AMOUNT']*.07 + p['PREMIUM_AMOUNT']
change_ph = propPh.sum() - p['PREMIUM_AMOUNT'].sum()
change_ph
```

```python
#Calculating the Change in premium amounts of Mobile and Travel insurance after premium discount by 10%
mt = am[(am['INSURANCE_TYPE'] == 'Mobile')| (am['INSURANCE_TYPE'] == 'Travel')]
mt1 = mt['PREMIUM_AMOUNT'] - mt['PREMIUM_AMOUNT']*.1
change_mt = mt1.sum() - mt['PREMIUM_AMOUNT'].sum()
change_mt
```

```python
total_pa = am['PREMIUM_AMOUNT'].sum()
total_pa
```

```python
sum_of_change_in_pa = change_ml + change_ph + change_mt
sum_of_change_in_pa
```

```python
total_change_in_pa = total_pa - sum_of_change_in_pa
total_change_in_pa
```

```python
#Answer
percent_change_in_pa = sum_of_change_in_pa/total_pa*100
percent_change_in_pa
```
